```python
import numpy as np
import matplotlib.pyplot as plt
import random


F = 1.6
plt.rcParams.update({
    "text.usetex": False, 
    "mathtext.fontset": "cm",
    "font.family": "serif",  
    "font.serif": ["cmr10"], 
    "font.size": 11*F,        
    "axes.labelsize": 11*F,   
    "axes.titlesize": 12*F,   
    "legend.fontsize": 10*F,  
    "xtick.labelsize": 10*F, 
    "ytick.labelsize": 10*F, 
    "figure.dpi": 300,     
    "savefig.bbox": 'tight',   
    "savefig.pad_inches": 0.1,  
    "pgf.texsystem": "pdflatex" 
})


a = 16807
c = 0
m = (2**31) - 1

sys_rng = random.SystemRandom()  
x_0 = sys_rng.randint(2**31, 2**40)



def LCG(N):
    rn = []
    xn = x_0
    for i in range(N):
        xn1 = (a*xn) % m
        rn.append(xn1/m)
        xn = xn1
    return rn

sample1 = LCG(1000)
sample2 = LCG(10000)



fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

fig.suptitle('The Minimal Standard')

ax1.hist(sample1, bins=50, density=True, alpha=0.6, color="blue")
ax1.plot([0,1], [1,1], color = 'red', label='Uniform (0,1) Distribution')

ax2.hist(sample2, bins=50, density=True, alpha=0.6, color="blue")
ax2.plot([0,1], [1,1], color = 'red', label='Uniform (0,1) Distribution')


ax1.set_title('n = 1000')
ax2.set_title('n = 10000')

ax1.set_xlabel('$LCG[2^{31}-1,16807,0]$')
ax2.set_xlabel('$LCG[2^{31}-1,16807,0]$')
ax1.set_ylabel('$Density$')
ax2.set_ylabel('$Density$')

ax1.legend()
ax2.legend()
plt.tight_layout()
plt.savefig('the_minimal_standard.pdf', 
            dpi=300, 
            format='pdf', 
            transparent=True,  
            bbox_inches='tight')


plt.show()

```
