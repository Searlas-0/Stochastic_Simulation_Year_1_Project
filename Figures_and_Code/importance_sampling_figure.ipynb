```python
import numpy as np
import matplotlib.pyplot as plt 

F = 1.6

plt.rcParams.update({
    "text.usetex": False, 
    "mathtext.fontset": "cm",
    "font.family": "serif",  
    "font.serif": ["cmr10"], 
    "font.size": 11*F,        
    "axes.labelsize": 11*F,   
    "axes.titlesize": 12*F,   
    "legend.fontsize": 10*F,  
    "xtick.labelsize": 10*F, 
    "ytick.labelsize": 10*F, 
    "figure.dpi": 300,     
    "savefig.bbox": 'tight',   
    "savefig.pad_inches": 0.1,  
    "pgf.texsystem": "pdflatex",
    "axes.formatter.use_mathtext": True
})

colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
    
np.random.seed(42)

samples = 20

def loss_function(x):
    return 0.9 * np.exp(-x**2 / 0.02) + 0.1 * np.exp(-x**2 / 2)

xs = np.linspace(-1, 1, 1000)

markers_uniform_x = np.random.uniform(-1, 1, samples)
markers_uniform_y = loss_function(markers_uniform_x)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

ax1.plot(xs, loss_function(xs))
ax1.scatter(markers_uniform_x, markers_uniform_y, color='red', marker='x', s=60, label='Uniform random sample')
ax1.set_title('Uniform Sampling of Guassian Function')
ax1.set_xticklabels([])
ax1.set_yticklabels([])

importance_samples = np.random.normal(0, 0.1, samples) 
importance_samples = np.clip(importance_samples, -1, 1)

ax2.plot(xs, loss_function(xs))
ax2.scatter(importance_samples, loss_function(importance_samples), color='red', marker='x', s=60, label='Normal p.d.f. random sample')
ax2.set_title('Importance Sampling of Guassian Function')
ax2.set_xticklabels([])
ax2.set_yticklabels([])

ax1.tick_params(axis='both', which='both', length=0)
ax2.tick_params(axis='both', which='both', length=0)
ax1.set_ylabel('$f(x)$')
ax2.set_ylabel('$f(x)$')
ax1.set_xlabel('$x$')
ax2.set_xlabel('$x$')

ax1.legend()
ax2.legend()
plt.tight_layout()
plt.savefig('importance_sampling_figure.pdf', 
            dpi=300, 
            format='pdf', 
            transparent=True,  
            bbox_inches='tight')
plt.show()
```
